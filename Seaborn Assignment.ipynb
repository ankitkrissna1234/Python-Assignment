Seaborn is a popular data visualization library in Python that provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots that can be created using Seaborn, along with their uses:

i) Scatter Plot:

Uses: It is used to visualize the relationship between two numerical variables. It helps to identify patterns, clusters, or trends in the data.
ii) Line Plot:

Uses: It is commonly used to display trends over time or continuous data. It helps to visualize the progression or change in a variable.
iii) Bar Plot:

Uses: It is useful for comparing categorical variables or showing the distribution of a single categorical variable. It is often used to visualize frequencies, proportions, or comparisons between different groups.                           
iv) Histogram:

Uses: It is used to represent the distribution of a single numerical variable. It helps to understand the shape, central tendency, and spread of the data.                                                                              
v) Box Plot:                            

Uses: It is used to visualize the distribution of numerical data and identify potential outliers. It provides information about the median, quartiles, and any extreme values in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
sns.boxplot(data=df, x="pclass", y="age")
plt.title("Box Plot of Age by Passenger Class")
plt.show()
```

```python
sns.boxplot(data=df, x="pclass", y="fare")
plt.title("Box Plot of Fare by Passenger Class")
plt.show()
```

```python
df = sns.load_dataset('diamonds')
```

```python
sns.histplot(data=df, x='price', hue='cut', multiple='stack')
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('flights')
```

```python
df_pivot = df.pivot('month', 'year', 'passengers')
```

```python
sns.heatmap(data=df_pivot, cmap='YlGnBu')
plt.title('Flights Dataset Heatmap')
plt.show()
```
