```python
import scipy.stats as stats

mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100

# Calculate the standard error
standard_error = std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1) * standard_error

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Print the confidence interval
print("Confidence Interval: [{:.2f}, {:.2f}]".format(lower_bound, upper_bound))
```

```python
import scipy.stats as stats

observed = [25, 18, 22, 11, 12, 12]  # Observed counts of each color
expected = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # Expected proportions of each color
sample_size = sum(observed)  # Total sample size

# Calculate the expected counts
expected_counts = [prop * sample_size for prop in expected]

# Perform the chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(observed, expected_counts)

# Print the test results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

# Check if the null hypothesis is rejected or not
alpha = 0.05
if p_value < alpha:
    print("The distribution of colors in the bag does not match the expected distribution.")
else:
    print("The distribution of colors in the bag matches the expected distribution.")
```

```python
import scipy.stats as stats
import numpy as np

# Create the contingency table
observed = np.array([[20, 15], [10, 25], [15, 20]])

# Perform the chi-square test of independence
chi2_stat, p_value, _, _ = stats.chi2_contingency(observed)

# Print the test results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

# Check if the null hypothesis is rejected or not
alpha = 0.05
if p_value < alpha:
    print("There is a significant association between the outcomes.")
else:
    print("There is no significant association between the outcomes.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set the degrees of freedom
df = 10

# Generate x values
x = np.linspace(0, 30, 1000)

# Generate y values (probability density function)
y = stats.chi2.pdf(x, df)

# Create the plot
plt.plot(x, y)

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
y_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='gray', alpha=0.3)

# Label the axes
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability Density Function')

# Add a title
plt.title('Chi-square Distribution (Degrees of Freedom = 10)')

# Display the plot
plt.show()
```

```python
import scipy.stats as stats
import numpy as np

# Create the contingency table
observed = np.array([[20, 15], [10, 25], [15, 20]])
expected = stats.contingency.expected_freq(observed)

# Perform the chi-square test for independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

# Print the test results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of freedom:", dof)

# Check if the null hypothesis is rejected or not
alpha = 0.05
if p_value < alpha:
    print("There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("There is no significant association between smoking status and lung cancer diagnosis.")
```
