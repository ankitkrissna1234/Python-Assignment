Matplotlib is a popular plotting library for Python that provides a wide variety of customizable and high-quality 2D and 3D plots. It is widely used in data visualization tasks and scientific computing. Matplotlib is highly versatile and can be used for creating basic plots, statistical visualizations, interactive visualizations, and much more.

Here are five common plots that can be created using the Pyplot module of Matplotlib:

i) Line plot: A line plot is used to represent data points connected by straight lines. It is commonly used to display trends or patterns over time or other continuous variables.

ii) Scatter plot: A scatter plot displays individual data points as markers on a two-dimensional plane. It is useful for visualizing the relationship between two variables and identifying any possible correlations.

iii) Bar plot: A bar plot (or bar chart) represents categorical data with rectangular bars of different heights or lengths. It is often used to compare and display data across different categories or groups.

iv) Histogram: A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins and displays the frequency or proportion of data points falling into each bin.

v) Pie chart: A pie chart represents data as a circle divided into sectors, where each sector represents a particular category and the area of each sector corresponds to the proportion of data it represents. It is commonly used to show the composition of a whole or the relative distribution of categories.

2)

A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane. It is particularly useful for visualizing the relationship between two variables and identifying any patterns or correlations.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

3)

The 'subplot()' function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots in a single figure, which can be helpful for comparing different datasets or displaying related plots together.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

A bar plot, also known as a bar chart or bar graph, is a type of plot that represents categorical data using rectangular bars. The length or height of each bar corresponds to the value or frequency of the category it represents. Bar plots are commonly used to compare and visualize data across different categories or groups

Bar plots are used for various purposes, including:

i) Comparing quantities: Bar plots allow you to easily compare the values or quantities associated with different categories. They make it visually clear which category has a larger or smaller value.

ii) Showing trends or patterns: Bar plots can display changes in values or frequencies over time or across different conditions. They can highlight trends or patterns in the data, making it easier to understand and interpret.

iii) Presenting rankings: Bar plots are useful for displaying rankings or ordering of categories based on their values. They provide a clear visual representation of the hierarchy or order among categories.

iv) Communicating insights: Bar plots are effective in conveying information and insights to an audience. They are visually appealing and easy to understand, making them suitable for presentations, reports, and data storytelling.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays summary statistics such as the median, quartiles, and potential outliers. Box plots are used to gain insights into the distribution, spread, and skewness of the data, as well as to compare multiple datasets.

The box plot consists of several components:

i) The box represents the interquartile range (IQR), which spans from the 25th percentile (Q1) to the 75th percentile (Q3) of the data. It visually shows the middle 50% of the data.                                                        
ii) Inside the box, a line or marker represents the median value, which is the 50th percentile.                        
iii) The whiskers extend from the box to show the range of the data, excluding any potential outliers.                 
iv) Any data points outside the whiskers are considered potential outliers and are displayed as individual points.

Box plots are used for various purposes, including:

i) Identifying skewness and outliers: Box plots provide information about the distribution of the data and help identify if it is skewed or if there are any potential outliers.                                                       
ii) omparing distributions: Box plots allow for easy visual comparison of multiple datasets, enabling the identification of differences in central tendency, spread, and skewness between different groups or categories.        
iii) Summarizing data: Box plots provide a concise summary of the dataset, highlighting key statistics such as the median, quartiles, and range.                                                                                          
iv) Visualizing variability: The length of the whiskers in a box plot represents the variability or spread of the data. A longer whisker indicates a greater spread, while a shorter whisker suggests a more compact distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data')
plt.ylabel('Values')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
